```python
train["DataType"] = "Titanic"
train
```

```python
train["id"] = range(0,891)
train
```

```python
train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
train[["SibSp","Parch","FamilySize"]]
```

```python
train["Nationality_FR"] = [f(i) for i in train["Embarked"]]
train["Nationality_UK"] = train["Embarked"].isin(["S","Q"])
train[["Embarked", "Nationality_FR", "Nationality_UK"]]
```

```python
train["Fare_Low"] = train["Fare"] < 20
train["Fare_Medium"] = (train["Fare"] >= 20) & (train["Fare"] < 100)
train["Fare_High"] = train["Fare"] >= 100

train[["Fare", "Fare_Low", "Fare_Medium", "Fare_High"]]
```

```python
train.loc[train["Embarked"] == "C", "Nationality"] = "FR"
train.loc[train["Embarked"] == "S", "Nationality"] = "UK"
train.loc[train["Embarked"] == "Q", "Nationality"] = "UK"
train[["Embarked", "Nationality"]]
```

```python
train.loc[train["Age"].isnull(), "Age"] = train["Age"].mean()
```
