```python
import csv
from bs4 import BeautifulSoup
import pandas as pd
```

```python
data = pd.read_csv('./survey/study_pre_survey.csv')
```

```python
data.head()
```

```python
data.drop(['available', 'studytime'], axis=1, inplace=True)
data.head()
```

# 위치

## 1. 설문결과 정리

```python
#출발 위치 / 귀가 위치를 구단위로 정리함. 

place_total = []

# 설문지에서 출발위치/귀가위치를 가져와서 하나의 리스트로 정리
check = ['comefrom(gu)','home(gu)']
for l in check:
    gus = data[l].values    #gus = list(data[l])  과 동일함
    for gu in gus:
        place_total.append(gu)


# 위치(구단위)별 카운트
place_set = set(place_total)
# print(place_set)
place_summary = {}
for place in place_set:
    place_summary[place] = place_total.count(place)

print(place_summary)
```

## 2. 이미지로 표시

### 서울시 지도 준비
![지도](./survey/begin.jpg)

```python
svg = open('./survey/Seoul_districts.svg','r').read()   #서울시 지도 가져오기
soup = BeautifulSoup(svg, 'lxml')
paths = soup.select('path')
colors = ['#F1EEF6', '#D4B9DA','#DD1C77', '#980043', '#dd1c1c']    # 색상 지정.  5단계로.. 
path_style = 'font-size:12px;fill-rull:nonzero;\
            stroke:#FFFFFF;stroke-opacity:1;stroke-width:0.1;stroke-miterlimit:4;stroke-dasharray:none; \
            stroke-linecap:butt;market-start:none;stroke-linejoin:bevel;fill:'
```

## 3. 설문조사 지도에 표시


```python
# 카운트 결과에 따라 지도 색상 변경

for p in paths:   # 서울시의 모든 구를 한번씩 p에 대입
    
    # 설문조사 결과를 count 에 저장한다. 설문조사 결과가 없으면 continue
    try: 
        count = place_summary[p['id']]
        # place_summary = {'Eunpyeong-gu': 1, 'Gwanak-gu': 2, ..... }
    except:
        continue
        
    if count >= 4:
        color_class = 4
    else:
        color_class = count

    color = colors[color_class]
    p['style'] = path_style + color
        
print(soup.prettify())
# 결과값을 .svg 파일로 저장 : place_in_seoup.svg
```
